```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
fares = pd.read_csv('https://raw.githubusercontent.com/Kaartik7/ISSC_competition-/pomegranate02-patch-1/to%20csv/transport%20fares.csv')
```

```python
fares
```

```python
new_fares = fares[fares['REF_DATE'] == '2015/2016']
```

```python
new_fares
```

```python
new_fares = new_fares[new_fares['Statistics'] == 'Per capita transport expenditures']
```

```python
new_fares
```

```python
plt.bar(new_fares['GEO'], new_fares['VALUE'],color='orange')
plt.title('Per Capita Transport Expenditure for Different Provinces in Canada')
plt.xlabel('Amount in Dollars')
plt.ylabel('Province')
plt.xticks(rotation=90)
plt.show()
```

```python
new_fares = fares[(fares['GEO'] == 'Ontario') & (fares['Statistics'] == 'Per capita transport expenditures')]
```

```python
new_fares
```

```python
plt.plot(new_fares['REF_DATE'], new_fares['VALUE'])
plt.scatter(new_fares['REF_DATE'], new_fares['VALUE'])
plt.title('Per Capita Transport Expenditure in Ontario from the year 2007 to 2016')
plt.xlabel('Year')
plt.ylabel('Amount in Dollars')
plt.xticks(rotation=90)
plt.show()
```
